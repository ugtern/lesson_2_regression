```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from numpy.linalg import inv
```

```python
x = np.linspace(1,10, num=10).reshape(-1, 1)
y = [1.5, 2.5, 3, 4.5, 12, 6.7, 7, 8.5, 14, 7]

plt.scatter(x, y)
plt.show()
```

```python
regression = LinearRegression().fit(x, y)
r2_score(y, regression.predict(x))
```

```python
y_log = np.log(y)

plt.scatter(x, y_log)
plt.show()
```

```python
regression = LinearRegression().fit(x, y_log)
r2_score(y_log, regression.predict(x))
```

```python
y_square = np.square(y)

plt.scatter(x, y_square)
plt.show()
```

```python
regression = LinearRegression().fit(x, y_square)
r2_score(y_square, regression.predict(x))
```

```python
y_sqrt = np.sqrt(y)

plt.scatter(x, y_sqrt)
plt.show()
```

```python
regression = LinearRegression().fit(x, y_sqrt)
r2_score(y_sqrt, regression.predict(x))
```

```python
y = np.array(y)
```

```python
y_SS = StandardScaler().fit_transform(y.reshape(-1, 1)).reshape(-1) # Z-score normalization
y_SS
```

```python
plt.scatter(x, y_SS)
plt.show()
```

```python
regression = LinearRegression().fit(x, y_SS)
r2_score(y_SS, regression.predict(x))
```

```python
MMS = MinMaxScaler().fit_transform(y.reshape(-1,1)).reshape(-1)
MMS
```

```python
plt.scatter(x, MMS)
plt.show()
```

```python
regression = LinearRegression().fit(x, MMS)
r2_score(MMS, regression.predict(x))
```
